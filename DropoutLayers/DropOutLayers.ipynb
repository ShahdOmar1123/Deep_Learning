```python
import pandas as pd
import numpy as ny
import tensorflow as tf
from tensorflow import keras
from keras import models,layers
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/sonar_dataset.csv',header=None)
data.sample(5)
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
data[60].value_counts()
```

```python
x=data.drop(60,axis=1)
y=data[60]
```

```python
y=pd.get_dummies(y,drop_first=True).astype(int)
y.sample(5)
```

```python
y.value_counts()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model=keras.Sequential([
    layers.Dense(60,input_dim=60,activation='relu'),
    layers.Dense(30,activation='relu'),
    layers.Dense(15,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,batch_size=8,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_predict=model.predict(x_test).reshape(-1)
print(y_predict[:10])
y_predict=ny.round(y_predict)
print(y_predict[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import  confusion_matrix , classification_report
print(classification_report(y_test,y_predict))
```

```python
#model with DropOut
modell=keras.Sequential([
    layers.Dense(60,input_dim=60,activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(30,activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(15,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
modell.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
modell.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
modell.evaluate(x_test,y_test)
```

```python
y_predictt=modell.predict(x_test).reshape(-1)
print(y_predictt[:10])
y_predictt=ny.round(y_predictt)
print(y_predictt[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import  confusion_matrix , classification_report
print(classification_report(y_test,y_predictt))
```
